```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from google.colab import files

# Загрузка данных
uploaded = files.upload()

# Получаем имя загруженного файла
file_name = list(uploaded.keys())[0]

# Прочитайте загруженный файл
data = pd.read_csv(file_name)
```

```python
# Реализация критерия Пирсона хи-квадрат
def chi_square_uniformity_test(data, column):
    observed_freq, bins = np.histogram(data[column], bins='auto')
    expected_freq = np.full_like(observed_freq, len(data) / len(observed_freq))

    chi_square_stat, p_value = stats.chisquare(f_obs=observed_freq, f_exp=expected_freq)

    return chi_square_stat, p_value

chi_square_stat, p_value = chi_square_uniformity_test(data, 'battery_power')
print(f"Chi-square Statistic: {chi_square_stat}, P-value: {p_value}")
```
