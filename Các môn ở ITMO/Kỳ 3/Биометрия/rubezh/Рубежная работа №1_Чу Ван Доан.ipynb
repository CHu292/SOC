Подсчитайте по городам (регионам) среднее и медианное значение, при необходимости округлите до целого методом round()

```python
data1 = data[data['ОКАТО'] == 'г. Кудымкар']
data2 = data[data['ОКАТО'] == 'г. Пермь']
print(f"г. Кудымкар: {str(round(data1['Значение'].mean()))}")
print(f"г. Пермь: {str(round(data2['Значение'].mean()))}")
print()
print(f"г. Кудымкар: {str(round(data1['Значение'].median()))}")
print(f"г. Пермь: {str(round(data2['Значение'].median()))}")
```

Выведите графическим способом полученные результаты по каждому городу в отдельности
